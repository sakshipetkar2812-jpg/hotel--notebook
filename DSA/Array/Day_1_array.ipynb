Problem 1: Print all elements of an array

Problem:
Given an array, print all its elements.

Example

Input: [1, 2, 3, 4]
Output: 1 2 3 4

```python
array=[1,2,3,4]
for i in array:
  print(i,end="")
```

Problem 2: Find the sum of elements

Problem:
Find the total sum of all elements in an array.

Example

Input: [1, 2, 3]
Output: 6

```python
array=[1,2,3]
sum=0
for i in array:
  sum+=i
print(sum)
```

Problem 3: Find maximum element

Problem:
Find the largest number in the array.

Example

Input: [2, 5, 1, 9, 3]
Output: 9

```python
array=[2,5,1,9,3]
max_var = array[0]

for i in array:
  if i>max_var:
    max_var=i

print(max_var)
```

Problem 4: Count even numbers

Problem:
Count how many even numbers are in the array.

Example

Input: [1, 2, 3, 4, 6]
Output: 3

```python
arr= [1,2,3,4,6]
count=0
for i in arr:
  if i%2==0:
    count=count+1
print(count)
```

Problem 5: Search an element (Linear Search)

Problem:
Check if a number exists in the array.

Example

Array: [10, 20, 30, 40]
Target: 30
Output: Found

```python
arr = [10, 20, 30, 40]
target = 30
found = False

for x in arr:
    if x == target:
        found = True
        break

if found:
    print("Found")
else:
    print("Not Found")

```
