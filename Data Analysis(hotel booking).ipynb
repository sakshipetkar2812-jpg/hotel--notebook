### Data Analysis and Visualization

```python
# Step-8: Barplot of the Reservation Status
cancelled_perc = df['is_canceled'].value_counts(normalize = True)
print(cancelled_perc)
```

```python
plt.figure(figsize= (5,4))
plt.title('Reservation status count')
plt.bar(['Not canceled','Canceled'],df['is_canceled'].value_counts(), edgecolor='k',width=0.7)
plt.show()
```

```python
# Step-9: Countplot for Reservation Status in differents Hotel
plt.figure(figsize =(8,4))
ax1 = sns.countplot(x='hotel',hue ='is_canceled',data=df,palette='Blues')


ax1.legend(['Not canceled','Canceled'],title='Status',loc='upper left',bbox_to_anchor=(1,1))

ax1.set_title('Reservation status in different hotels',size=20)

ax1.set_xlabel('Hotel')
ax1.set_ylabel('Number of reservations')
plt.tight_layout()
plt.show()
```

```python
resort_hotel = df[df['hotel']== 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel = df[df['hotel']== 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
# Step-10: plot for Average Daily Rate in City and Resort Hotel
plt.figure(figsize=(20,8))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize=30)
plt.plot(resort_hotel.index,resort_hotel['adr'],label='Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label='City Hotel')
plt.legend(fontsize=20)
plt.show()
```

```python
# Step-11: Countplot for Reservation status per Month

df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize=(16,8))
ax2 = sns.countplot(x='month',hue='is_canceled',data=df, palette='bright')
legeng_labels,_ =ax2. get_legend_handles_labels()
ax2.legend(bbox_to_anchor=(1,1))
plt.title('Reservation status per month',size=20)
plt.xlabel('month')
plt.ylabel('Number of Reservations')
plt.legend(['not canceled','canceled'])
plt.show()
```

```python
# Step-12: Barplot of the ADR per Month
plt.figure(figsize=(15,8))
plt.title('ADR per month',fontsize=30)
sns.barplot(x='month',y='adr', hue='month',data = df[df['is_canceled']==1].groupby('month')['adr'].sum().reset_index(), palette='viridis')
plt.legend([],[],frameon=False)
plt.show()
```

```python
# Step-13: Piechart for Top 10 Countries with reservation canceled
canceled_data= df[df['is_canceled']==1]
top_10 = canceled_data['country'].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.title('Top 10 Countries with reservation canceled')
plt.pie(top_10,autopct='%.2f',labels =top_10.index)
plt.show()
```

```python
not_canceled_data = df[df['is_canceled']==0]
```

```python
canceled_df_adr = canceled_data.groupby('reservation_status_date')[['adr']].mean()
canceled_df_adr.reset_index(inplace=True)
canceled_df_adr.sort_values('reservation_status_date',inplace=True)

not_canceled_df_adr=not_canceled_data.groupby('reservation_status_date')[['adr']].mean()
not_canceled_df_adr.reset_index(inplace=True)
not_canceled_df_adr.sort_values('reservation_status_date',inplace=True)

plt.figure(figsize=(20,6))
plt.title('Average Daily Rate')
plt.plot(not_canceled_df_adr['reservation_status_date'],not_canceled_df_adr['adr'],label='not_canceled')
plt.plot(canceled_df_adr['reservation_status_date'],canceled_df_adr['adr'],label='canceled')
plt.legend()
plt.show()


```

```python
canceled_df_adr = canceled_df_adr[(canceled_df_adr['reservation_status_date']>'2016') & (canceled_df_adr['reservation_status_date']<'2017-09')]

not_canceled_df_adr = not_canceled_df_adr[(not_canceled_df_adr['reservation_status_date']>'2016') & (not_canceled_df_adr['reservation_status_date']<'2017-09')]
```

```python
# Step-14: Plot of Average Daily Rate for year 2016 t0 2017
plt.figure(figsize=(20,6))
plt.title('Average Daily Rate',fontsize=30)
plt.plot(not_canceled_df_adr['reservation_status_date'],not_canceled_df_adr['adr'],label='not_canceled')
plt.plot(canceled_df_adr['reservation_status_date'],canceled_df_adr['adr'],label='canceled')
plt.legend()
plt.show()
```
